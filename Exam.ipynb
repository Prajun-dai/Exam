```python
def check_for_coconut(string_list):
 
  return 'coconut' in string_list


# Example 1: 'coconut' is in the list
list_with_coconut = ["apple", "banana", "coconut", "date"]
result1 = check_for_coconut(list_with_coconut)
print(f"List 1 contains 'coconut': {result1}")

# Example 2: 'coconut' is not in the list
list_without_coconut = ["grape", "kiwi", "mango"]
result2 = check_for_coconut(list_without_coconut)
print(f"List 2 contains 'coconut': {result2}")

# Example 3: Different casing does not match
list_wrong_case = ["Coconut", "pineapple"]
result3 = check_for_coconut(list_wrong_case)
print(f"List 3 contains 'coconut' (case-sensitive): {result3}")
```

```python
def main():
    strings = []

    while True:
        s = input("Enter a string (press Enter to finish): ")
        if s == "":
            break
        strings.append(s.lower())

    strings.sort()

    print("\nStrings in alphabetical order (lower case):")
    for s in strings:
        print(s)


if __name__ == "__main__":
    main()
```
